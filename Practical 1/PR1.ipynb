```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))
```

```python
def tanh(x):
    return np.tanh(x)
```

```python
def relu(x):
    return np.maximum(0,x)
```

```python
def leakyRelu(x,alpha=0.01):
    return np.where(x>0,x,x*alpha)
```

```python
def softmax(x):
    return np.exp(x)/np.sum(np.exp(x))
```

```python
x=np.linspace(-5,5,100)
y_sigmoid=sigmoid(x)
y_tanh=tanh(x)
y_relu=relu(x)
y_lrelu=leakyRelu(x)
y_soft=softmax(np.array([x]))
```

```python
plt.figure(figsize=(12,8))
plt.subplot(2,3,1)
plt.plot(x,y_sigmoid,label='Sigmoid')
plt.subplot(2,3,2)
plt.plot(x,y_tanh)
plt.subplot(2,3,3)
plt.plot(x,y_relu)
plt.subplot(2,3,4)
plt.plot(x,y_lrelu)
plt.subplot(2,3,5)
plt.plot(x,y_soft.T)
```
